```python
print('Ni le taux de charge ni le taux de couverture no sont pas renseignés dans ces 20 premières lignes…')
print('Je charge les 1500000 premières lignes…')
elec = pd.read_csv('..\CSV\eco2mix-regional-cons-def.csv', sep = ';', nrows = 1500000)
```

```python
print(elec.iloc[elec['TCO Hydraulique (%)'].isna().argmin(), :])
print('Ce sont les données de la ligne où la colonne TCO Hydraulique est renseignée pour la première fois.')
```

```python
print('Une autre façon de trouver cette ligne est d’utiliser la méthode “first_valid_index')
print('Cette méthode permet d’être utilisée sur plusieurs colonnes à la fois.')
elec.iloc[elec.iloc[:, 15:-1].first_valid_index(), :]
```
